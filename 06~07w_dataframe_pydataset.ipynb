###결측값을 채우는 함수 fillna()

```python
import pandas as pd
import numpy as np

df  = pd.DataFrame([[np.nan,2,np.nan,0],
                    [3,4,np.nan,1],
                    [np.nan,np.nan,np.nan,np.nan],
                    [np.nan,3,np.nan,4]],
                    columns=list("ABCD"))
df
```

```python
df.fillna(10)
```

```python
df
```

```python
df.fillna('missing',axis=0)
```

```python
df.fillna('missing',limit=1)
```

```python
df.fillna('missing',limit=1,axis=0)
```

```python
df.fillna('missing',limit=2)
```

```python
df.fillna('missing',limit=1,axis=1)
```

```python
from IPython import display
display
```

```python
from IPython.display import HTML, display
def display_side_by_side(*args):
    html_string = ''
    for df_arg in args:
        html_string += df_arg.to_html() + "&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"
    display(HTML(html_string))

display_side_by_side(df, df.fillna("missing",limit=1,axis=1))
```

```python
display_side_by_side(df, df.fillna("missing",limit=2,axis=1))
```

```python
values = {'A':100,'B':200,'C':300,'D':400}
df.fillna(values)
```

```python
values
```

```python
display_side_by_side(df, df.fillna(values))
```

```python
display_side_by_side(df, df.fillna(values,limit=2))
```

```python
s = pd.Series([100,200,400],index=list('ACD'))
s
```

```python
print(s)
```

```python
display_side_by_side(df, df.fillna(s,limit=1))
```

```python
display_side_by_side(df, df.fillna(s,limit=2))
```

```python
df2 = pd.DataFrame(-np.ones((4,4)),columns = list('ABCD'))
```

```python
display_side_by_side(df, df.fillna(df2))
```
