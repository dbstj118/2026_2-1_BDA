```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000) #1000개의 랜덤 값 추출
y = np.random.randint(-100,100,1000) #1000개의 랜덤 값 추출

mask1 = abs(x) > 50 #x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) > 50 #y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1 + mask2] #mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] #mask1과 mask2 중 하나라도 만족하는 값 저장

plt.scatter(x,y,c=x,cmap='jet',alpha=0.4)
plt.colorbar()
plt.show()
```

###plt.subplot(), plt.subplots(), fig.add_subplot() 그림제목

```python
import matplotlib.pyplot as plt

#전체 제목 - 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목',fontsize=14, y=1.02)

#subplot1 - .생성 직후 바로 title 지정
plt.subplot(2,3,1)
plt.plot([1,4,9,16])
plt.title('제곱수')

#subplt5
plt.subplot(2,3,6)
plt.plot([1,2,3,4])
plt.title('선형')

#subplot2
plt.subplot(2,3,2)
plt.plot([6,5,2,1])
plt.title('감소')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,3,figsize=(7,4))
fig.set_size_inches(6,3)

ax[0,1].plot([1,4,9,16])
ax[1,0].plot([1,2,3,4])
ax[1,2].plot([6,5,2,1])

# 전체 제목 - Figure 레벨
fig.suptitle('전체 제목',fontsize=14,fontweight='bold')

# 각 subplot 제목 - Axes 레벨
ax[0,0].set_title('(0,0) 빈 플롯')
ax[0,1].set_title('(0,1) 제곱수')
ax[0,2].set_title('(0,2) 빈 플롯')
ax[1,0].set_title('(1,0) 선형')
ax[1,1].set_title('(1,1) 빈 플롯')
ax[1,2].set_title('(1,2) 감소')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,4))

ax1 = fig.add_subplot(231)
ax1.plot([1,4,9,16])
ax1.set_title('제곱수')

ax2 = fig.add_subplot(233)
ax2.plot([1,2,3,4])
ax2.set_title('선형')

ax3 = fig.add_subplot(235)
ax3.plot([6,5,2,1])
ax3.set_title('감소')

fig.suptitle('전체 제목',fontsize=14)
plt.tight_layout()
plt.show()
```
